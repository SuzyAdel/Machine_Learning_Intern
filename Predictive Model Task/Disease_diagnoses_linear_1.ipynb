# Moduling

```python
data.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
train = data.drop(['Disease', 'ALT', 'AST', 'Mean Corpuscular Volume'], axis=1)
test = data['Disease']
```

```python
x_train, x_test, y_train, y_test= train_test_split(train,test,test_size=0.3, random_state=2)
```

```python
# Encode non-numerical target data
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.fit_transform(y_test)
```

```python
reger= LinearRegression()
```

```python
reger.fit(x_train, y_train_encoded)
```

```python
predictions = reger.predict(x_test)
```

```python
predictions
```

```python
reger.score(x_test,y_test_encoded)
```

the accuracy is very poor as its modeled  using linear regression while it is actually logistic regression as the values are not contineous
